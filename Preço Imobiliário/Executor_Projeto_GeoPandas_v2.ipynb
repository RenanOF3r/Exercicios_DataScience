# Executor do Projeto GeoPandas (Alura)

Baixa o `Projeto_completo.ipynb` do repositório da Alura, executa todas as células (incluindo gráficos)
e salva uma cópia com todas as saídas em **`Projeto_completo_executado.ipynb`**.


```python
# Parâmetros
INPUT_NB_URL = "https://raw.githubusercontent.com/alura-cursos/curso_geopandas/main/Projeto_completo.ipynb"
INPUT_NB_PATH = "Projeto_completo.ipynb"
OUTPUT_NB_PATH = "Projeto_completo_executado.ipynb"
print("Notebook de entrada:", INPUT_NB_URL)
print("Notebook executado será salvo em:", OUTPUT_NB_PATH)
```

```python
import sys, subprocess
def pip_install(*packages):
    cmd = [sys.executable, "-m", "pip", "install", "-q"] + list(packages)
    print("Instalando:", " ".join(packages))
    subprocess.check_call(cmd)

# Papermill para executar notebooks e Requests para baixar o .ipynb de origem
pip_install("papermill", "requests")
# Obs.: o notebook original instala geopandas.
```

```python
import requests, json
r = requests.get(INPUT_NB_URL)
r.raise_for_status()
nb_json = r.json()
with open(INPUT_NB_PATH, "w", encoding="utf-8") as f:
    json.dump(nb_json, f)
print("Baixado:", INPUT_NB_PATH)
```

```python
import papermill as pm
print("Executando notebook. Isso pode levar alguns minutos...")
pm.execute_notebook(
    INPUT_NB_PATH,
    OUTPUT_NB_PATH,
    kernel_name="python3",
)
print("Concluído! Arquivo gerado:", OUTPUT_NB_PATH)
```
