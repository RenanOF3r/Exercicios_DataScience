# Executor do Projeto GeoPandas (Alura)

Este notebook baixa o projeto **Projeto_completo.ipynb** do repositório da Alura e o **executa inteiro**, salvando uma cópia com **todas as saídas** (gráficos inclusos) em `Projeto_completo_executado.ipynb`.

**Como usar (Colab ou local):**
1. *Opcional (Colab):* Conecte e, se quiser, monte o Google Drive para salvar o resultado lá.
2. Rode as células na ordem (Ctrl/Cmd+F9 ou Runtime → Run all).
3. Ao final, faça o download do arquivo `Projeto_completo_executado.ipynb`.


```python
# Parâmetros
INPUT_NB_URL = "https://raw.githubusercontent.com/alura-cursos/curso_geopandas/main/Projeto_completo.ipynb"
INPUT_NB_PATH = "Projeto_completo.ipynb"
OUTPUT_NB_PATH = "Projeto_completo_executado.ipynb"

print("Notebook de entrada:", INPUT_NB_URL)
print("Notebook executado será salvo em:", OUTPUT_NB_PATH)
```

```python
import sys, subprocess
def pip_install(*packages):
    cmd = [sys.executable, "-m", "pip", "install", "-q"] + list(packages)
    print("Instalando:", " ".join(packages))
    subprocess.check_call(cmd)

# Papermill para executar notebooks e Requests para baixar o .ipynb de origem
pip_install("papermill", "requests")
# Observação: o notebook original já instala geopandas quando executado.
```

```python
import requests, json
r = requests.get(INPUT_NB_URL)
r.raise_for_status()
nb_json = r.json()
with open(INPUT_NB_PATH, "w", encoding="utf-8") as f:
    json.dump(nb_json, f)
print("Baixado:", INPUT_NB_PATH)
```

```python
import papermill as pm
print("Executando notebook. Isso pode levar alguns minutos...")
pm.execute_notebook(
    INPUT_NB_PATH,
    OUTPUT_NB_PATH,
    kernel_name="python3",
)
print("\nConcluído! Arquivo gerado:", OUTPUT_NB_PATH)
```

```python
# (Opcional) Se estiver no Google Colab, faça o download automático do resultado
try:
    from google.colab import files
    files.download(OUTPUT_NB_PATH)
except Exception as e:
    from pathlib import Path
    print("Se não estiver no Colab, baixe manualmente o arquivo:", Path(OUTPUT_NB_PATH).resolve())
```
